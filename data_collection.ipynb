### Calculate Time to Deliver:
- time_to_delivery_from_shipment
- time_to_delivery_from_order_approved
- time_to_delivery_from_order_purchased

````python
df_orders['order_delivered_carrier_date'] = df_orders['order_delivered_carrier_date'].astype('datetime64[ns]')
df_orders['order_delivered_customer_date'] = df_orders['order_delivered_customer_date'].astype('datetime64[ns]')
df_orders['order_approved_at'] = df_orders['order_approved_at'].astype('datetime64[ns]')
df_orders['order_purchase_timestamp'] = df_orders['order_purchase_timestamp'].astype('datetime64[ns]')
df_orders
````

````python
def calculate_time_diff(df_start, df_end):
    return (df_end - df_start) / pd.Timedelta(days=1)
````

````python
df_orders['time_to_delivery_from_shipment'] = calculate_time_diff(df_orders['order_delivered_carrier_date'], df_orders['order_delivered_customer_date'])
df_orders['time_to_delivery_from_order_approved'] = calculate_time_diff(df_orders['order_approved_at'], df_orders['order_delivered_customer_date'])
df_orders['time_to_delivery_from_order_purchased'] = calculate_time_diff(df_orders['order_purchase_timestamp'], df_orders['order_delivered_customer_date'])
df_orders
````

````python
import geopy.distance
````

````python
def calculate_distance_between_coords(lat1, lng1, lat2, lng2):
    coords_1 = (lat1, lng1)
    coords_2 = (lat2, lng2)

    return geopy.distance.geodesic(coords_1, coords_2).km
````

````python
df_order_seller_geo_customer_merged["shipment_distance"] = df_order_seller_geo_customer_merged.apply(lambda row: calculate_distance_between_coords(row['customer_lat'], row['customer_lng'], row['seller_lat'], row['seller_lng']), axis=1)
df_order_seller_geo_customer_merged
````

````python
df_final = df_order_seller_geo_customer_merged.sort_values("order_purchase_timestamp", ascending=True, ignore_index=True)
df_final = df_final.drop_duplicates(ignore_index=True)
df_final = df_final.drop(columns=["customer_id", "seller_lat", "seller_lng", "customer_lat", "customer_lng"])
df_final
````

````python
duplicated_ids = df_final[df_final["order_id"].isin(df_final["order_id"][df_final["order_id"].duplicated()])].sort_values("order_id")
df_final = df_final[~df_final['order_id'].isin(duplicated_ids["order_id"].values.tolist())]
df_final
````

````python
df_final = df_final.drop(columns=["product_id", "geolocation_zip_code_prefix_x", "geolocation_zip_code_prefix_y"])
df_final
````

````python
df_final.to_csv("./dataset.csv", index=False)
````
